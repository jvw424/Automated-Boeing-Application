```python


# This is the decision tree for answering personal history questions

def ansQuestion(question):
    if not 'how many years' in question:
        if 'willing to travel' in question:
            return 'Yes'
        if 'have a bachelor' in question and 'computer science' in question:
            return 'Yes'
        if 'ability to obtain' in question and 'clearance' in question:
            return 'Yes'
        if 'able to obtain' in question and 'clearance' in question:
            return 'Yes'
        if '18 years of age' in question:
            return 'Yes'
        if 'highest level' in question and 'clearance' in question:
            return 'None of the Above'
        if 'held an active' in question and 'clearance' in question:
            return 'No'
        if "bachelor" in question and 'degree' in question:
            return 'Yes'
        if 'agile' in question:
            return 'Yes'
        if '10 year experience' in question:
            return 'No'
        if 'git' in question or 'gitlab' in question or 'github' in question:
            return 'Yes'
        if 'ui/ux' in question:
            return 'Yes'
        if 'u.s. person' in question or 'u.s. citizen' in question or 'lawful permanent resident' in question:
            return 'Yes'
        if 'experience' in question:
            return 'Yes'
    else:
        if 'python' in question:
            return 'More than 3 years but less than 5 years'
        if 'git' in question:
            return 'More than 3 years but less than 5 years'
        if 'nosql' in question:
            return 'More than 3 years but less than 5 years'
        if 'object-oriented' in question:
            return 'More than 3 years but less than 5 years'
        if 'higher education' in question:
            return 'More than 3 years but less than 5 years'
        if 'java' in question:
            return 'More than 1 year but less than 3 years'
        if 'javascript' in question:
            return 'More than 1 year but less than 3 years'
        if 'agile' in question:
            return 'More than 1 year but less than 3 years'
        if 'react' in question:
            return 'More than 1 year but less than 3 years'
        if 'node.js' in question:
            return 'More than 1 year but less than 3 years'
        if 'css' in question:
            return 'More than 1 year but less than 3 years'
        if 'c/c++' in question:
            return 'More than 1 year but less than 3 years'
        if 'software development lifecyle' in question or 'sdlc' in question:
            return 'More than 1 year but less than 3 years'
        if 'real time' in question:
            return 'More than 1 year but less than 3 years'
        if 'linux' in question:
            return 'Less than 1 year'
        if 'docker' in question:
            return 'Less than 1 year'
        if 'kubernetes' in question:
            return 'Less than 1 year'
        if 'podman' in question:
            return 'Less than 1 year'
        if 'mongodb' in question:
            return 'Less than 1 year'
        if 'c++' in question:
            return 'Less than 1 year'
        if 'vue' in question:
            return 'Less than 1 year'
        if 'angular' in question:
            return 'Less than 1 year'
        if 'cyber security' in question:
            return 'Less than 1 year'
        else:
            return 'Less than 1 year'
        


def getDate():
    today = date.today()

    if today.day <= 9:
        day = str(0) + str(today.day)
    else:
        day = str(today.day)
    if today.month <= 9:
        month = str(0) + str(today.month)
    else:
        month = str(today.month)

    year = str(today.year)
    
    return month+day+year
        
    
```
